```python
import pandas as pd
import numpy as np
```

# Loading the data

```python
#loading the dataframe 
df = pd.read_csv("Insurance_complaints_tdi.csv")

#taking a look at the data
df.head()
```

## Exploring the data

```python
#Number of columns and rows 
print("Shape of the dataframe")
print(df.shape)


```

```python
#Taking look at the index
df.index

#The columns 
df.columns

#Info on dataframe
df.info()
```

```python
summary_data = []

for col in df.columns:
    
    #the entire column as a list
    entire_col = df[col]    
    
    #Adding column name
    col_name = col
    
    #Getting all unique values 
    unique_vals = list(entire_col.unique())
    
    #Getting the missing values 
    missing_vals = entire_col.isnull().sum()
    
    col_summary = {
        "Column Name": col_name,
        "Unique Values": unique_vals,
        "Missing Rows": missing_vals,
        #"Mean": mean_val,
        #"Median": median_val
    }
    
    summary_data.append(col_summary)
    
    
summary_df = pd.DataFrame(summary_data)
summary_df.head()    
```
